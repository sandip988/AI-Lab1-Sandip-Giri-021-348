```python
import random

class VacuumCleaner:
    def __init__(self):
        self.width = 4
        self.height = 4
        self.grid = [[1 if random.random() < 0.3 else 0 for _ in range(4)] for _ in range(4)]
        self.pos = [0, 0]  # Agent position

    def display(self):
        for y in range(self.height):
            row = ''
            for x in range(self.width):
                if [x, y] == self.pos:
                    row += 'A' if self.grid[y][x] == 1 else 'a'
                else:
                    row += str(self.grid[y][x])
            print(row)
        print()

    def is_dirty(self, x, y):
        return self.grid[y][x] == 1

    def clean(self, x, y):
        self.grid[y][x] = 0

    def is_all_clean(self):
        return all(cell == 0 for row in self.grid for cell in row)

    def act(self):
        x, y = self.pos
        # Clean if current cell is dirty
        if self.is_dirty(x, y):
            self.clean(x, y)
            return 'Clean'

        # Find nearest dirty cell
        min_dist = float('inf')
        target = None
        for i in range(self.height):
            for j in range(self.width):
                if self.is_dirty(j, i):
                    dist = abs(j - x) + abs(i - y)
                    if dist < min_dist:
                        min_dist = dist
                        target = (j, i)

        # Move toward dirty cell or randomly
        if target:
            tx, ty = target
            if tx > x and x < self.width - 1:
                self.pos[0] += 1
                return 'Right'
            elif tx < x and x > 0:
                self.pos[0] -= 1
                return 'Left'
            elif ty > y and y < self.height - 1:
                self.pos[1] += 1
                return 'Down'
            elif ty < y and y > 0:
                self.pos[1] -= 1
                return 'Up'
        # Random move if no dirty cells or can't move toward target
        move = random.choice(['Up', 'Down', 'Left', 'Right'])
        if move == 'Up' and y > 0:
            self.pos[1] -= 1
        elif move == 'Down' and y < self.height - 1:
            self.pos[1] += 1
        elif move == 'Left' and x > 0:
            self.pos[0] -= 1
        elif move == 'Right' and x < self.width - 1:
            self.pos[0] += 1
        return move

    def run(self, max_steps=100):
        steps = 0
        while not self.is_all_clean() and steps < max_steps:
            print(f"Step {steps + 1}:")
            self.display()
            action = self.act()
            print(f"Action: {action}")
            steps += 1
        print("Final state:")
        self.display()
        print("Goal achieved!" if self.is_all_clean() else "Max steps reached.")

if __name__ == "__main__":
    cleaner = VacuumCleaner()
    cleaner.run()
```
